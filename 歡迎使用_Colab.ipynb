```
import matplotlib.pyplot as plt

id = input("請輸入學號")
name = input("請輸入姓名")
cy = input("目前學年度")
se = input("請輸入英文成績")
cse = input("請輸入英文學分數")
sm = input("請輸入數學成績")
csm = input("請輸入數學學分數")
ss = input("請輸入體育成績")
css = input("請輸入體育學分數")
sc = input("請輸入國文成績")
csc = input("請輸入國文學分數")
ey = id[:3]
dc = id[3:5]
sn = id[5:]
gy = str(int(ey)+3)
cg = str(int(cy)-int(ey)+1)
avgs_float = ((int(se))+int(sm)+int(ss)+int(sc))/4 # avgs 可以是小數
acs_float = (int(se)*int(cse))+(int(sm)*int(csm))+(int(ss)*int(css))+(int(sc)*int(csc)) #  acs 可以是小數
total_credits = int(cse)+int(csm)+int(css)+int(csc) # 計算總學分數
cavgs_float = acs_float/total_credits # 除法運算

plt.figure(figsize=(9, 6)) # 寬9高6
subjects = ['ENGLISH', 'MATH', 'SPORT', 'CHINESE']
scores = [int(se), int(sm), int(ss), int(sc)]
bars = plt.bar(subjects, scores, color='skyblue', edgecolor='black', alpha=0.8)
plt.title("SCORE",fontsize=11,pad=20)
plt.xlabel("SUBJECT",fontsize=12)
plt.ylabel("Score",fontsize=12)
plt.ylim(0, 100)
plt.grid(axis='y',linestyle='--',alpha=0.6)

print(f"系所代號;{dc}")
print(f"預計畢業時間;{gy}年")
print(f"當前年級;{cg}年級")
print(f"平均分數;{avgs_float:.2f}分") # 兩位小數
print(f"加權平均分數;{cavgs_float:.2f}分") # 兩位小數
plt.show()
```
